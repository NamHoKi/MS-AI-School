```python
import pandas as pd
purchase = pd.read_csv('purchase.csv')
```

```python
purchase
```

```python
from sklearn.preprocessing import StandardScaler
x = purchase[['Age','EstimatedSalary']]

sc_purchase = StandardScaler()
x_sc= sc_purchase.fit_transform(x)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

plt.figure(figsize=(12,12))

for k in range(1,7):
  km = KMeans(n_clusters=k)
  km.fit(x_sc)
  ids = km.predict(x_sc)

  plt.subplot(3,2, k)
  plt.tight_layout()

  plt.title(f'k={k}')
  plt.xlabel('age')
  plt.ylabel('salary')

  plt.scatter(x_sc[:,0], x_sc[:,1], c=ids)

plt.show()
```

```python
k=2
km_2 = KMeans(n_clusters=k)
km_2.fit(x_sc)

km_df_2 = purchase.copy()
ids = km_2.predict(x_sc)

km_df_2['cluster'] = ids
km_df_2.head()
```

```python
km_df_2[km_df_2['Purchased'] != km_df_2['cluster']]
```
