```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,100)
ax.plot(x, np.sin(x))    # 특정 축에 그리기
```

```python
# 색상 변경
plt.plot(x, np.sin(x-0), color='red')
plt.plot(x, np.sin(x-1), color='m')  # magenta
plt.plot(x, np.sin(x-2), color='k')  # black
plt.plot(x, np.sin(x-3), color='0.75')  # 0-1 사이의 회색조를 표현한다.
plt.plot(x, np.sin(x-4), color='#FFDD44')  #  rgb 16진수
plt.plot(x, np.sin(x-5), color=(1.0, 0.2, 0.3)) # rgb 값을 튜플로 표현 
```

```python
# 선 변경
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')


plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--c')
plt.plot(x, x+2, '-.k')
plt.plot(x, x+3, ':r')
```

```python
# 플롯 조정하기: 축 경계
plt.plot(x, np.sin(x))

plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x, np.sin(x))
plt.xlim(0,10)
plt.ylim(1.5,-1.5)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5])
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal') # x,y축 크기 같음
```

```python
# 플롯에 레이블 붙이기
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.legend()
```

```python
# 산점도 출력

x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='k')
```

```python
rnd = np.random.RandomState(0)
for marker in ['o',',','x','+','v','<','>','s','d']:
    plt.plot(rnd.rand(5), rnd.rand(5), marker, label='marker={0}'.format(marker))
    plt.legend()
    plt.xlim(0,1.8)
```
